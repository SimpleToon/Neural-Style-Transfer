```python
from ImageProcessor import ImageProcessor
from AdaIN import AdaIN
```

```python
decoder= "decoder-adain.pth"
encoder = "vgg.pth"

image_path = "content/avril.jpg"
style_path = "style/sketch.png"
```

```python
adainModel = AdaIN(encoder, decoder)
adainModel.setup()
adainModel.fit(image_path, [style_path])
adainModel.pipeline([1])
adainModel.displayImages()
adainModel.evaluate()
```

```python
adainModel = AdaIN(encoder, decoder, colorPreservation= "Pixel")
adainModel.setup()
adainModel.fit(image_path, [style_path])
adainModel.pipeline([1])
adainModel.displayImages()
adainModel.evaluate()
```

```python
adainModel2 = AdaIN(encoder, decoder, colorPreservation= "Histogram")
adainModel2.setup()
adainModel2.fit(image_path, [style_path])
adainModel2.pipeline([1])
adainModel2.displayImages()
adainModel2.evaluate()
```
