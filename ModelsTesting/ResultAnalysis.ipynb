```python
import numpy as np
import pandas as pd
# Result from my google form: https://forms.gle/pnY9Tnni76nQRgHNA
```

```python
df = pd.read_csv("Neural Style Transfer - Model (Responses) - Form Responses 1.csv")
df
```

```python
df = df.loc[:,df.columns.str.startswith("Tracking")]
df
```

```python
flatten_array = df.values.flatten()
flatten_array
```

```python
remove_nan_arr = [i for i in flatten_array if isinstance(i, str)]
remove_nan_arr
```

```python
result = {}
for item in remove_nan_arr:
    splitArr = item.split("_")
    
    if splitArr[0] == "A":
        if splitArr[1] in result:
            result[splitArr[1]] += 1
        else: 
            result[splitArr[1]] = 1
    if splitArr[0] == "B":
        if splitArr[3] in result:
            result[splitArr[3]] += 1
        else: 
            result[splitArr[3]] = 1
result
```

```python
total_win_rate =  pd.DataFrame(
    result.items(),
    columns=["Model", "Wins"]
)
total_win_rate["Total Win Rate %"] = round(total_win_rate["Wins"]/total_win_rate["Wins"].sum() * 100,2)
total_win_rate
```

```python
pairwise_result = {}
for item in remove_nan_arr:
    splitArr = item.split("_")
    concatString = splitArr[1] + splitArr[2] + splitArr[3]

    #Initalise key
    if concatString not in pairwise_result:
        pairwise_result[concatString] = [0,0]
        
    if splitArr[0] == "A":
        pairwise_result[concatString][0] += 1 

    if splitArr[0] == "B":
        pairwise_result[concatString][1] += 1 

pairwise_result
```

```python
pairwise_win_rate = pd.DataFrame.from_dict(
    pairwise_result,
    orient="index",
    columns=["Model1_wins", "Model2_wins"]
)
pairwise_win_rate
```
